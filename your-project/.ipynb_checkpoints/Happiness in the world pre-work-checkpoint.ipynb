```python
#importing data bases

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Who suicide database (until 2015)

```python
suic = pd.read_csv('data\who_suicide_statistics.csv')
suic.head()
```

```python
suic['country'].unique()
```

### Percentage of suicides per country

```python
#percentage of suicides over population per country

suic['percentage']= (suic['suicides_no'] / suic['population'])*100
```

### Dataset with only the desired countries (for all the years)

```python
#Data set only with the desired countries

country_list = ['Australia','Brazil','Japan','Spain','United States of America','United Kingdom']

df = suic[suic['country'].isin(country_list)]

df.head()
```

### Save this as pd df

```python
df.to_csv(r'data\suicides_7countries.csv')
```

### Gridplot of the general data frame (with the desired countries)

```python
g = sns.PairGrid(df, hue="country", diag_sharey=False, corner=True)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
```

```python
sns.pairplot(df, hue="country")
```

```python
fg,ax=plt.subplots(figsize=(15,4))

ax = sns.barplot(x="country", y="percentage", hue="age", data=df )

plt.savefig('suicides_per_age_and_country.png')
```

## World Happiness report

```python
happy = pd.read_csv('data\world-happiness-report.csv')
happy.head()
```

### Dataset with only the desired countries

```python
#Data set only with the desired countries

country_list = [
    'Australia','Brazil','Japan','Spain','South Africa','United States','United Kingdom','Finland','Nigeria','Vietnam','Canada','Argentina']

len(country_list)
```

```python
happy_countries = happy[happy['Country name'].isin(country_list)]

happy_countries.head()
```

```python
happiness = happy_countries.drop(['Generosity','Positive affect','Negative affect'],axis=1)
```

```python
happiness.to_csv(r'data\happiness.csv')
```

```python
num = happiness.select_dtypes(exclude='object')

num

sns.heatmap(num.corr(),annot=True,cmap="YlGnBu", linewidths=.5)
plt.savefig('matrix.png')
```

```python
sns.set_theme(style="darkgrid")

f = sns.jointplot(x="Social support", y="Healthy life expectancy at birth", data=happiness)
```

## Data Bank suicides database (2005-2019)

```python
sui_databank = pd.read_csv('data\Completa.csv')
sui_databank.head(25)
```

```python
finalsuic = sui_databank.drop(['Series Name','Series Code','Country Code'],axis=1)
finalsuic.head()
```

```python
finalsuic.columns = ['Country','2005','2006','2007','2008','2009','2010','2011','2012','2013','2014','2015','2016','2017','2018','2019']
finalsuic.head()
```

### Melting 

```python
final_suic = pd.melt(finalsuic, id_vars=['Country'])
final_suic = final_suic.dropna()

final_suic.columns = ['Country','Year','Suicide rate (per 100K people)']
```

```python
final_suic.head(50)
```

### Joining databases

```python
happiness.head()
```

```python
happiness.columns = ['Country','Year','Life Ladder','GDP per capita','Social support','Life expectancy','Freedom','Corruption']
```

```python
happiness = happiness.dropna()
```

```python
happiness.head(50)
```

```python
final_suic.head()
```

```python
final_suic['new'] = final_suic['Country'] + '-' + final_suic['Year']
final_suic.head()
```

```python
happiness['new'] = happiness['Country'] + '-' + happiness['Year'].astype(str)
happiness.head()
```

```python
result = happiness.merge(final_suic,left_on='new', right_on='new')
```

```python
result.head(50)
```

```python
# Dropping and rename columns

result = result.drop(['new','Country_y','Year_y'],axis=1)
```

```python
result.columns = ['Country','Year','Life Ladder','GDP per capita','Social support','Life expectancy','Freedom','Corruption','Suicide rate (per 100K people)']
```

```python
result[result['Country']=='Vietnam']
```

```python
result.to_csv(r'data\FINAL.csv',index=False)
```
