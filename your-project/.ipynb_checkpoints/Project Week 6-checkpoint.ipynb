```python
original_df.loc[original_df["Currency Code"] == "Australian Dollar", ['Unit Price','Product Cost Amount','Tax Amount']] = original_df.loc[original_df["Currency Code"] == "Australian Dollar", ['Unit Price','Product Cost Amount','Tax Amount']] * 0.78
original_df.loc[original_df["Currency Code"] == "Canadian Dollar", ['Unit Price','Product Cost Amount','Tax Amount']] = original_df.loc[original_df["Currency Code"] == "Canadian Dollar", ['Unit Price','Product Cost Amount','Tax Amount']] * 0.81
original_df.loc[original_df["Currency Code"] == "Deutsche Mark", ['Unit Price','Product Cost Amount','Tax Amount']] = original_df.loc[original_df["Currency Code"] == "Deutsche Mark", ['Unit Price','Product Cost Amount','Tax Amount']] * 0.617
original_df.loc[original_df["Currency Code"] == "French Franc", ['Unit Price','Product Cost Amount','Tax Amount']] = original_df.loc[original_df["Currency Code"] == "French Franc", ['Unit Price','Product Cost Amount','Tax Amount']] * 0.1839

original_df.loc[original_df["Currency Code"].isin(["United Kingdom Pound","GBP"]), ['Unit Price','Product Cost Amount','Tax Amount']] = original_df.loc[original_df["Currency Code"].isin(["United Kingdom Pound","GBP"]), ['Unit Price','Product Cost Amount','Tax Amount']] * 1.39
```

```python
original_df
```

```python
clean_df = original_df.drop(columns=['Currency Code', 'Sale Amount', 'Promotion Discount Pct', 'Promotion Discount Amount', 'Delivery Cost Amount', 'Order Quantity', 'GUID'])
```

```python
clean_df
```

```python
clean_df["Sales Territory ID"].value_counts()
```

```python
clean_df["Sales Territory ID"] = clean_df["Sales Territory ID"].str.replace("^ ", "")
clean_df["Sales Territory ID"] = clean_df["Sales Territory ID"].str.replace(".", "")
clean_df["Sales Territory ID"] = clean_df["Sales Territory ID"].str.replace("Europa", "Europe")
clean_df["Sales Territory ID"] = clean_df["Sales Territory ID"].str.replace("Pacifics", "Pacific")
```

```python
clean_df["Profit"] = round(clean_df["Unit Price"] - clean_df["Product Cost Amount"] - clean_df["Tax Amount"], 2)
```

```python
clean_df
```

```python
monthly_profit = clean_df.groupby(clean_df["Order Date"].dt.strftime('%m'))["Profit"].sum()
```
